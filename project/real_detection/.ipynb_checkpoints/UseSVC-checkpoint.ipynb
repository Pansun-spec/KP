```python
import numpy as np
import cv2
from matplotlib import pyplot as plt
from skimage.io import imread, imsave
from sklearn.svm import LinearSVC
```

```python
def flatten_img(img):
    x, y, c = img.shape
    return img.reshape((x * y, c))
```

```python
img = imread('res.jpg')
plt.imshow(img)
```

```python
svc = LinearSVC()
```

```python
# Assign parameters from FindHolesSVC to fit model
svc.coef_ = np.asarray([[ 0.05897537, 0.03026204, -0.11213523]])
svc.intercept_ = np.asarray([0.10923962])
svc.classes_ = np.asarray([0., 1.])
```

```python
%%time
pred_svc = svc.predict(flatten_img(img))
```

```python
img_mask = pred_svc.reshape(img.shape[0], img.shape[1]).astype(np.uint8)
plt.figure(figsize=(16, 9))
plt.imshow(img_mask, cmap='gray')
```
